```python
#importing libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# importing library
path = r"C:\your\base\directory"
input_dir = os.path.join(path, '02 Data', 'Prepared Data')
```

```python
pickle_file = os.path.join(input_dir, 'ords_prods_cust_merge.pkl')
```

```python
ords_prods_cust_merge = pd.read_pickle(pickle_file)
```

```python
#creating histogram for 'order_hour_of_day'
hist = ords_prods_cust_merge['order_hour_of_day'].plot.hist(bins = 24)
```

The busiest hours of the day are 8am-4pm

```python
#creating a bar chart for "loyalty_flag"
bar = ords_prods_cust_merge['loyalty_flag'].value_counts().plot.bar()
```

```python
#creating a sample of the dataframe to check difference
sampled_df = ords_prods_cust_merge.sample(n=100000, random_state=1)
```

```python
#group by hour of the day and calculate the mean expenditure
hourly_expenditure = sampled_df.groupby('order_hour_of_day')['prices'].mean().reset_index()
```
